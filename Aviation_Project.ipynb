### Data Preparation
### Handling missing values
Since we are working with a large dataset we will need to drop unnecessary columns and handle missing values. 
We also have to fix the naming of the certain columns with wrong naming conventions.

```python
#renaming columns
aviation_df.rename(columns={"Event.Id": "Event_Id", 
                            "Investigation.Type":"Investigation_Type",
                            "Accident.Number":"Accident_Number", 
                            "Event.Date":"Event_Date", 
                            "Airport.Code":"Airport_Code", 
                            "Airport.Name":"Airport_Name",
                            "Injury.Severity":"Injury_Severity",
                            "Aircraft.damage":"Aircraft_damage",
                            "Aircraft.Category":"Aircraft_Category",
                            "Registration.Number":"Registration_Number",
                            "Amateur.Built":"Amateur_Built",
                            "Number.of.Engines":"Number_of_Engines",
                            "Engine.Type":"Engine_Type",
                            "FAR.Description":"FAR_Description",
                            "Purpose.of.flight":"Purpose_of_flight",
                            "Air.carrier":"Air_carrier",
                            "Total.Fatal.Injuries":"Total_Fatal_Injuries",
                            "Total.Serious.Injuries":"Total_Serious_Injuries",
                            "Total.Minor.Injuries":"Total_Minor_Injuries",
                            "Total.Uninjured":"Total_Uninjured",
                            "Weather.Condition":"Weather_Condition",
                            "Broad.phase.of.flight":"Broad_phase_of_flight",
                            "Report.Status":"Report_Status",
                            "Publication.Date":"Publication_Date",}, inplace=True)
```

```python
#check renamed columns
aviation_df.head()
```

```python
#drop columns
columns_to_drop= ['Latitude','Longitude','Airport_Code','Accident_Number','Purpose_of_flight','Registration_Number',
                   'FAR_Description','Schedule','Air_carrier','Broad_phase_of_flight','Report_Status','Publication_Date','Number_of_Engines']
aviation_df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
#check columns
aviation_df.head()
```

We will use specific values to fill in the null values in the remaining columns to avoid skewing the data using mean or median.

For the null values that are numerical we will use 0

For the null values that are categorical we will use "Unavailable"

```python
#replacing null values in categorical data with unavailable
aviation_df.fillna({'Location': "Unavailable"}, inplace=True)
aviation_df.fillna({'Country': "Unavailable"}, inplace=True)
aviation_df.fillna({'Airport_Name': "Unavailable"}, inplace=True)
aviation_df.fillna({'Injury_Severity': "Unavailable"}, inplace=True)
aviation_df.fillna({'Aircraft_damage': "Unavailable"}, inplace=True)
aviation_df.fillna({'Aircraft_Category': "Unavailable"}, inplace=True)
aviation_df.fillna({'Make': "Unavailable"}, inplace=True)
aviation_df.fillna({'Model': "Unavailable"}, inplace=True)
aviation_df.fillna({'Amateur_Built': "Unavailable"}, inplace=True)
aviation_df.fillna({'Engine_Type': "Unavailable"}, inplace=True)
aviation_df.fillna({'Weather-Condition': "Unavailable"}, inplace=True)

#replacing null numerical values with zero
aviation_df.fillna({'Total_Fatal_Injuries': 0}, inplace=True)
aviation_df.fillna({'Total_Serious_Injuries': 0}, inplace=True)
aviation_df.fillna({'Total_Minor_Injuries': 0}, inplace=True)
aviation_df.fillna({'Total_Uninjured': 0}, inplace=True)

aviation_df.head()
```

### Data Analysis
We will carry out exploratory data analysis for the cleaned data. This will allow us to find patterns within attributes through visualisations.

```python
#import necessary libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Top 10 Aircarft Makes with Highest Fatalities

```python
#Due to a high number of rows we will get total injuries by make 
# get total fatalities by aircraft make
grouped_data = aviation_df.groupby('Make')['Total_Fatal_Injuries'].sum().sort_values(ascending=False).head(10)

x = grouped_data.index  # top 10 aircraft makes
y = grouped_data.values  # fatalities

# bar chart
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='blue')
plt.xlabel('Aircraft Makes')
plt.ylabel('Number of Fatalities')
plt.title('Top 10 Aircraft Makes with Highest Fatalities')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

The aircraft make with the highest number fatalities is Cesna. This indicates that Cesna has had more accidents over the years. 

The aircraft with the least amount of fatalities is Airbus Industrie. This indicates that Airbus Industrie has had the least number of accidents over the years

### Top 10 Countries with the Highest Fatalities

```python
#Due to a high number of rows we will get total injuries by countries
# get total fatalities by aircraft make
grouped_data = aviation_df.groupby('Country')['Total_Fatal_Injuries'].sum().sort_values(ascending=False).head(10)

x = grouped_data.index  # Top 10 countries
y = grouped_data.values  # Corresponding fatalities

# bar chart plot
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='blue')
plt.xlabel('Countries')
plt.ylabel('Number of Fatalities')
plt.title('Top 10 Countries with the Highest Fatalities')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

The united states have the highest number of accidents geographically. This indicates that aviation started earlier in USA hence higher aircraft movements which results to more incidents recorded.

### Accident rate pattern over the years

```python
#extract year from datetime
aviation_df['Event_Date'] = pd.to_datetime(aviation_df['Event_Date'])
aviation_df['Year'] = aviation_df['Event_Date'].dt.year

# accidents by year
accidents_per_year = aviation_df['Year'].value_counts().sort_index()

# line chart plot
plt.figure(figsize=(10, 6))
plt.plot(accidents_per_year.index, accidents_per_year.values, marker='o', linestyle='-', color='skyblue')
plt.title('Accidents Rate Pattern over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

There were less recorded incidents from 1950 to the start of 1970s because the use of aircrafts was not a common means of transport. The 1980s recorded the highest number of accidents since more aircarfts strated being used. From the 1990s the number of accidents start to decrease due to improved technology.
